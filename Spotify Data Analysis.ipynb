```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#IMPORTING THE DATASET
df_tracks = pd.read_csv("C:\\tracks.csv")
df_tracks.head()
```

```python
#NULL VALUES
print(df_tracks.info())
```

```python
df_tracks.isnull().sum() #TOTAL NUMBER OF MISSING VALUES IN EACH COLUMNS
```

```python
#TOP 10 LEAST POPULAR SONGS
sorted_df = df_tracks.sort_values(by = ['popularity'], ascending = True).head(10)
sorted_df
```

```python
#STATISTICAL DATA ABOUT THE DATASET
df_tracks.describe()
```

```python
#TOP 10 MOST POPULAR SONGS
most_popular = df_tracks.sort_values(by= ['popularity'], ascending = False).head(10)
most_popular
```

```python
df_tracks.set_index('release_date' , inplace = True)
df_tracks
```

```python
#CONVERTING THE INDEX TO DATETIME FORMAT
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
#COVERTING DURATION IN MILLISECONDS TO SECONDS
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks['duration'].head()
```

```python
#CORRELATION BETWEEN VARIABLE
corr_df = df_tracks.drop(['key','mode','explicit'], axis = 1).corr(method ="pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, 
                      annot= True,
                      fmt=".1g",
                      vmin=-1,
                      vmax=1,
                      center = 0,
                      cmap="inferno",
                      linewidths=1,
                      linecolor = "Black")
heatmap.set_title("Correlation HeatMap Between Variables", fontsize = 14, weight = "bold")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation =90)
```

```python
#ACCOUTISNESS-ENERGY HAVE STRONG NEGATIVE COORELATION
#POPULARITY-LOUDNESS MODERATELY STRONG POSITIVE COORELATION
#DANCEABILITY-VALENCE MODERATELY STRONG POSITIVE COORELATION
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#REGRESSION PLOT B/W LOUDNESS & ENERGY
plt.figure(figsize=(10,8))
sns.regplot(data = sample_df,y='loudness',x='energy', color = "c",line_kws = {"color": "red"})
plt.title("Loudness vs Energy Correlation", fontsize = 14, weight ="bold")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.set_style('whitegrid')
sns.regplot(data = sample_df,y='popularity',x='acousticness', color = "b",line_kws = {"color": "black"})
plt.title("Popularity vs Acousticness Correlation", fontsize = 14, weight ="bold")
plt.show()
```

```python
#GETTING THE INITIAL VALUES OF THE RELEASE_DATE COLUMN USING 'GET_LEVEL_VALUES'
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
#NUMBER OF SONGS PER YEAR SINCE 1922
sns.displot(years, discrete=True, aspect=2,height=5,kind="hist")
plt.title("Number of Songs per year", fontsize= 14, weight="bold")
plt.xlabel("Years",fontsize=14)
plt.ylabel("Number of Songs", fontsize=14,)
plt.show()
```

```python
total_dr = df_tracks.duration
fig_dim  = (18,7)
fig,ax = plt.subplots(figsize=fig_dim)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False)
plt.title("Year vs Duration", fontsize=20,weight="bold")
plt.xticks(rotation=90)
plt.show()
```
